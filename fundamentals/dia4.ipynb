```python
# Challenge 2

# Imprime el menu de opciones
def print_menu():
    print(f'''
-----------------------
        MENU
1. Crear datos nuevos.
1. Modificar datos.
3. Eliminar datos.
4. Mostrar tarjeta.
          
0. Cerrar programa.
-----------------------
''')

# Agrega una nueva clave y valor
def crear_dato_nuevo(tarjeta):
    # Solicita la nueva clave
    nueva_clave = input(f'Ingrese una nueva clave...\n{print_menu()}')

    # verifica si la clave existe en la tarjeta
    if nueva_clave in tarjeta.keys():
        print(f'La clave {nueva_clave} ya existe. Puede modificar o eliminar la clave.')
        return
    
    # Solicita el nuevo valor
    nuevo_valor = input('Ingrese el nuevo valor...')
    
    # Guarda la nueva clave y su valor en la tarjeta
    tarjeta[nueva_clave] = nuevo_valor

    # print(tarjeta)
    return tarjeta

# Modifica la clave ingresada
def modificar_dato(tarjeta):
    # Solicita la clave a modificar
    clave_a_modificar = input(f'Ingrese clave a modificar...\n{print_menu()}')

    # verifica si la clave existe en la tarjeta
    if not clave_a_modificar in tarjeta.keys():
        print(f'La clave {clave_a_modificar} no existe. Puede crear una nueva la clave.')
        return
    
    # Solicita el nuevo valor
    nuevo_valor = input('Ingrese el nuevo valor...')
    
    # Guarda el nuevo valor en la clave ingresada
    tarjeta[clave_a_modificar] = nuevo_valor

    # print(tarjeta)
    return tarjeta

# Elimina un elemento
def eliminar_dato(tarjeta):
    # Solicita la clave a eliminar
    clave_a_eliminar = input(f'Ingrese clave a eliminar...\n{print_menu()}')

    # verifica si la clave existe en la tarjeta
    if not clave_a_eliminar in tarjeta.keys():
        print(f'La clave {clave_a_eliminar} no existe. Puede crear una nueva la clave.')
        return
    
    # elimina el elemente
    del tarjeta[clave_a_eliminar]

    # print(tarjeta)
    return tarjeta

# Retorna un numero entero
def get_option():
    while True:
        try:
            return int(input(f'Seleccione una opcion...\n{print_menu()}'))
        except:
            print('Opcion incorrecta... Por favor, ingrese un numero.')

# Imprime el contenido de la tarjeta
def print_tarjeta(tarjeta):
    for k, v in tarjeta.items():
        print(f'{k}: {v}')

# Inicializa un contenedor de datos de tipo diccionario
# PD.: se utiliza para guardar la informacion proporcionada por el usuario
tarjeta = {}

# Variable auxiliar para multiplicar el caracter 'x' veces
len_caracter = 50

# Imprime menu inicial
print_menu()

while True:
    # Obtiene un numero que representa las opciones
    opcion = get_option()

    # Verifica la eleccion del usuario
    if opcion == 0:
        print('=' * len_caracter)
        print_tarjeta(tarjeta)
        break
    elif opcion == 1:
        crear_dato_nuevo(tarjeta)
    elif opcion == 2:
        modificar_dato(tarjeta)
    elif opcion == 3:
        eliminar_dato(tarjeta)
    elif opcion == 4:
        print_tarjeta(tarjeta)
    else:
        print_menu()
```
